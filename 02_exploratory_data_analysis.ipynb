```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
attrition_dept = pd.crosstab(df["Department"], df["Attrition"])
attrition_dept
```

```python
attrition_dept.plot(kind="bar", figsize=(8,5))
plt.title("Attrition by Department")
plt.ylabel("Employees")
plt.show()
```

```python
attrition_gender = pd.crosstab(df["Gender"], df["Attrition"])
attrition_gender
```

```python
attrition_gender.plot(kind="bar")
plt.title("Attrition by Gender")
plt.show()
```

```python
attrition_ot = pd.crosstab(df["OverTime"], df["Attrition"])
attrition_ot
```

```python
attrition_ot.plot(kind="bar")
plt.title("Attrition by Overtime")
plt.show()
```

```python
travel = pd.crosstab(df["BusinessTravel"], df["Attrition"])
travel
```

```python
travel.plot(kind="bar")
plt.title("Attrition by Business Travel")
plt.show()
```

```python
df.boxplot(column="MonthlyIncome", by="Attrition", figsize=(6,5))
plt.title("Monthly Income vs Attrition")
plt.suptitle("")
plt.show()
```

```python
df.boxplot(column="Age", by="Attrition", figsize=(6,5))
plt.title("Age vs Attrition")
plt.suptitle("")
plt.show()
```

```python
corr = df.select_dtypes(include="number").corr()

corr
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

plt.imshow(corr, cmap="coolwarm", aspect="auto")

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)

plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")

plt.show()
```
